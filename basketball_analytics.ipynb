```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#creating the plot function

def myplot(data,playerlist = Players):

    Col = {"KobeBryant":"Black","JoeJohnson":"Red","LeBronJames":"Green","CarmeloAnthony":"Blue","DwightHoward":"Magenta","ChrisBosh":"Black","ChrisPaul":"Red",\
          "KevinDurant":"Green","DerrickRose":"Blue","DwayneWade":"Magenta"}

    Mrk =  {"KobeBryant":"s","JoeJohnson":"o","LeBronJames":"^","CarmeloAnthony":"D","DwightHoward":"s","ChrisBosh":"o","ChrisPaul":"^",\
          "KevinDurant":"D","DerrickRose":"s","DwayneWade":"o"} 

    for name in playerlist:
        plt.plot(data[Pdict[name]],c=Col[name],ls='--',marker=Mrk[name],ms='7',label=name)
    plt.legend(loc='upper left' ,bbox_to_anchor=(1,1))
    plt.xticks(list(range(0,10)),Seasons,rotation='vertical')
    plt.show()         
```
